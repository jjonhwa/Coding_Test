```python
from collections import deque

def solution(begin, target, words):
    answer = 0
    queue = deque([(begin, 0)])
    
    visits = {}
    for w in words:
        visits[w] = 0
    
    while queue:
        word, count = queue.popleft()
        visits[word] = count
        
        # 현재 선택한 단어와 한 글자 차이만 나는 알파벳 탐색
        for w in words:
            cnt = 0
            for a, b in zip(word, w):
                if a != b:
                    cnt += 1

            # 한 글자만 차이나는 경우
            if cnt == 1:

                # 한 번도 탐색하지 않은 단어이거나 /혹은/ 다른 경로가 더 짧을 경우 초기화
                if visits[w] == 0 or visits[word] < visits[w]:
                    visits[w] = count + 1
                    queue.append((w, count + 1))
     
    if target not in visits:
        return 0
    else:
        return visits[target]
```

- 본 문제는 탐색할 단어의 개수, 각 단어의 길이가 적고, 짧기 때문에 **완전탐색**을 통해 문제를 해결할 수 있다.
- 그 과정에서, dfs 혹은 bfs를 활용하여 조건에 맞는 값이 제 자리에 들어갈 수 있도록 코딩을 진행한다.

1. 탐색할 값들을 한 번에 하나씩 큐에 삽입
2. 큐에 삽입된 단어를 꺼내어 words의 단어들과 비교
3. 한 글자 차이만 나는 단어인지 아닌지 탐색
4. 한 글자 차이만 날 경우, 이미 탐색한 단어인지 확인 / 혹은 / 새롭게 탐색한 경로가 더 짧은 경로인지 확인
5. 2~4를 반복하며, queue에 값이 없을 때까지 반복.
