```python
from collections import defaultdict

def solution(genres, plays):
    answer = []

    genre_dict = defaultdict(list)
    genre_score = defaultdict(int)
    
    for i, (genre, play) in enumerate(zip(genres, plays)):
        genre_dict[genre].append((genre, play, i))
        genre_score[genre] += play
        
    for key, value in sorted(genre_score.items(), key = lambda x: x[1], reverse = True):
        sorted_genre = sorted(genre_dict[key], key = lambda x: (x[1], -x[2]), reverse = True)
        if len(sorted_genre) == 1:
            answer.append(sorted_genre[0][2])
        else:
            answer.append(sorted_genre[0][2])
            answer.append(sorted_genre[1][2])

    return answer
```

1. 속한 노래가 많이 재생된 장르를 먼저 수록한다.
    - 재생수를 count하기 위한 dictionary
    - 장르, 재생수, index를 가지기 위한 dictionary를 만든다.
2. 재생 수 순서대로 장르를 for loop 순환
3. 재생 수 순서대로 장르를 돌면서, 재생 수, 인덱스 번호가 낮은 노래 부터 수록한다.
    - 이 때, 곡이 한 곡 밖에 없을 경우 한 곡만 수록
    - 그렇지 않을 경우, 상위 2곡만 수록한다.
