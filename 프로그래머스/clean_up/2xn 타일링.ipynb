```python
def solution(n):
    dp = [0] * n
    dp[0], dp[1] = 1, 2
    
    for i in range(2, n):
        value = (dp[i-1] + dp[i-2]) % 1000000007
        dp[i] = value
        
    return dp[n-1]
```

- 1000000007으로 나눈 나머지를 return 한다.
    - 위의 조건에 의하여, n이 60000에 가까워질수록 시간초과가 발생할 수 있다.
    - 모든 값을 1000000007으로 나눈 나머지만을 메모이제이션 함으로서 시간을 더욱 단축할 수 있다.
    - 즉, 정답 부분에서 값의 나머지를 취하는 것이 아니라 매 value마다 나머지를 취함으로서 시간을 더 단축할 수 있다.
