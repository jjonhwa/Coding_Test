```python
from collections import deque, Counter

# 1번 노드로부터 각 노드 사이의 거리를 갱신한다.
n = 6
vertex = [[3, 6], [4, 3], [3, 2], [1, 3], [1, 2], [2, 4], [5, 2]]	

def solution(n, edge):
    graph = [[] for _ in range(n+1)]
    for v in edge:
        a, b = v
        graph[a].append(b)
        graph[b].append(a)

    # bfs를 통해서 연결된 노드까지의 거리를 갱신한다.
    queue = deque([1])

    # 각 node에서부터 1번 노드까지의 거리를 담는 dictionary 초기화
    distance = {}
    for i in range(1, n+1):
        if i == 1:
            distance[i] = 0
        else:
            distance[i] = 20_001

    # bfs로 값 갱신
    while queue:
        node = queue.popleft()

        # 거리 갱신
        # "다음노드-1번노드" 사이의 거리가 "현재노드-1번노드" 사이의 거리보다
        # 멀다면, 값을 갱신 후 queue에 값을 삽입
        # 그렇지 않을 경우, pass
        for _next in graph[node]:
            if distance[_next] > distance[node] + 1:
                distance[_next] = distance[node] + 1
                queue.append(_next)

    # 가장 먼 거리를 구하고, 그 거리를 취하는 값의 개수를 구한다.
    max_value = max(distance.values())
    counter_dict = Counter(distance.values())

    return counter_dict[max_value]
```

4. 최대값의 개수 count
